```python
import pandas as pd
import numpy as np
import math
from sklearn.neighbors import KNeighborsClassifier

data=pd.read_csv("train.csv")

train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
testy=pd.read_csv("gender_submission.csv")
test["Survived"]=testy.Survived
##initial training set
test.head()
```

```python
##creates a numeric nparray from training set
def createnumerics(data):
    data.head()
    type(data)
    ##dropping passenger ID column
    data=data.drop('PassengerId',1)
    ##dropping ticket number
    data=data.drop('Ticket',1)
    
    ## getting titles of passengers
    title=[]
    
    for i in range(data.shape[0]):
        title.append(data['Name'][i].split(",")[1].replace(".","").split(" ")[1])
    
    for i in range(len(title)):
        if title[i] == 'Ms':
            title[i] = 'Miss'
        elif title[i]=='Mme':
            title[i]='Mrs'
            
    title.count("Mr")+title.count("Mrs")+title.count("Master")+title.count("Miss")
    common=['Mr','Miss','Mrs','Master']
    for i in range(len(title)):
        if title[i] not in common:
            title[i]='rare'
    title
    # converting title into numerics
    """
    title_num=[]
    for i in title:
        if i == "Mr":
            title_num.append(1)
        elif i =="Miss":
            title_num.append(2)
        elif i=="Mrs":
            title_num.append(3)
        elif i=="Master":
            title_num.append(4)
        else:
            title_num.append(5)    
    """
    title_num=pd.factorize(title)
    title_num=list(title_num[0])
    
    title_num
    # sex to numerics    
    """
    sex_num=[]
    for i in data["Sex"]:
        if i =="male":
            sex_num.append(1)
        else:
            sex_num.append(0)
    """
    sex_num=pd.factorize(data["Sex"])
    sex_num=list(sex_num[0])
    
    sex_num
    
        
    ##embarked to numerics
    
    emb=data.Embarked.fillna('S')
    emb=list( )
    emb.count('S')
    
    embark_num=pd.factorize(data["Embarked"])
    embark_num=list(embark_num[0])
    
    embark_num    
    ##Analyze Fare
    
    fare=list(data.Fare)
    Fare=[]
    for i in fare:
        if not math.isnan(i):
            Fare.append(i)
    fare=data.Fare.fillna((int)(np.median(Fare)))   
    fare=[i/10 for i in fare]
    fare=[int(i) for i in fare]
    
    fare
    
    ##age
    Age=[]
    age=list(data.Age)
    for i in age:
        if not math.isnan(i):
            Age.append(i)
    age=data.Age.fillna((int)(np.median(Age)))
    age=[int(i)for i in age]
    
    age 
    
    ##SIblings & Spouses
    sibsp=list(data.SibSp)
    
    ##Parent - Children
    parch=list(data.Parch)
    ## Pclass
    pclass=list(data.Pclass)
    
    ## result
    y=list(data.Survived)
    
    ##NP-array of all obtained numerics.
    numerics=np.array([pclass,title_num,sex_num,age,sibsp,parch,fare,embark_num])    
    numerics=np.transpose(numerics)
    return (numerics,y)

##numerics of the training set
(X,Y)=createnumerics(train)
print(X)
print("\n\n")
print(Y)

```

```python
"""
implements KNN classifier from skl and returns accuracy
"""
def KNN_skl(train,test,n):
    (trX,trY)=createnumerics(train)
    (teX,teY)=createnumerics(test)
    y=np.array(teY)
    bleh=KNeighborsClassifier(n_neighbors=3)
    bleh.fit(trX,trY)
    Y=bleh.predict(teX)
    return(np.average(y==Y))


print(KNN_skl(train,test,3))

%timeit (KNN_skl(train,test,3))
```

```python
## Lazy classification K-Nearest Neighbors classification
"""
K - Nearest Neighbours Classifier(My Implementation)
"""
n=5
def KNN(train,test,n):
    (trX,trY)=createnumerics(train)
    (teX,teY)=createnumerics(test)
    Y=[]
    
    for i in teX:
        dist=[]
        for j in range(len(trX)):
            dist.append(((sum((i-trX[j])**2)),j))
            dist.sort()
        ind=[]
        for i in range(n):
            ind.append(dist[i][1])
        votes=0
        for i in ind:
            votes=votes+trY[i]
        if votes>n//2:
            Y.append(1)
           # np.append(trX,i)
           # trY.append(1)
        else:
            Y.append(0)
         #   np.append(trX,i)
          #  trY.append(0)
    q=np.array(Y)
    w=np.array(teY)
    return(np.average(q==w))    
##returns accuracy of predictions in test data, after noting from train data
print(KNN(train,test,3))

```
